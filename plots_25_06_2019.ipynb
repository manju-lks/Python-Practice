```
#iris dataset
iris = pd.read_csv("Iris.csv")
```

```
iris.head()
```

```
# Histogram
plt.figure(figsize=(15,10))
plt.hist(iris['SepalLengthCm'], bins=15, color='orange')
plt.xlabel("Sepal Length in cm")
plt.ylabel("Frequency")
plt.title("Histogram showing the frequency of Sepal Lengths of Iris")
plt.show()
```

```
# boxplot according species
iris.boxplot(by='Species', column=['SepalLengthCm'],grid=False)
```

```
# Scatter plot
plt.figure(figsize=(15,10))
plt.scatter(iris['SepalLengthCm'],iris['PetalWidthCm'], marker='^', c='r', label = "Sepal")
plt.scatter(iris['PetalLengthCm'],iris['PetalWidthCm'], marker='o', c='g', label = "Petal")
plt.xlabel("Sepal and Petal length in cm")
plt.ylabel("Petal Width in cm")
plt.title("Correlation between Length and Width of sepal and petal of Iris")
plt.legend()
plt.show()
```

```
# violin plot
plt.figure(figsize=(15,10))
plt.violinplot(iris['SepalLengthCm'])
```

```
# violin plot with seaborn
import seaborn as sns

sns.violinplot(x=iris['Species'], y=iris['SepalLengthCm'])
```

```
# line plot
plt.plot(iris["SepalWidthCm"])
```

```
# Diabetes 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
diabetes = pd.read_csv("pima-indians-diabetes.csv", header=None)
```

```
diabetes.head()
```

```
diabetes.columns=['n_pregnant', 'glu_conc', 'bp', 'tst', 'insulin', 'bmi', 'dpf', 'age', 'diabetes?']
```

```
diabetes.head()
```

```
#histogram

fig, ax = plt.subplots(1,3, sharey=True, tight_layout=True)
ax[0].hist(diabetes['n_pregnant'],bins=10, color='red')
ax[0].set_xlabel('Pregnancy')
ax[1].hist(diabetes['glu_conc'],bins=10, color='blue')
ax[1].set_xlabel('Glucose Concentration')
ax[2].hist(diabetes['bp'],bins=10, color='yellow')
ax[2].set_xlabel('Blood Pressure')
plt.show()
```

```
#boxplot
plt.figure(figsize=(15,10))
boxplot_data = diabetes[['n_pregnant', 'glu_conc', 'bp']]
boxplot_data.boxplot()
```

```
# scatter plot
plt.figure(figsize=(15,10))
plt.scatter(diabetes['glu_conc'],diabetes['insulin'], marker='p', c='g', label = "Glucose Concentration")
plt.scatter(diabetes['age'],diabetes['insulin'], marker='>', c='b', label = "Age")
plt.xlabel("glucose concentration and age")
plt.ylabel("insulin")
plt.title("Correlation between Glucose Concentration and age with Insulin")
plt.legend()
plt.show()
```

```
#Violin plot
plt.violinplot(diabetes['insulin'])
```

```
# violin plot with seaborn
import seaborn as sns

sns.violinplot(x=diabetes['diabetes?'], y=diabetes['insulin'])
```

```
#line plot
plt.plot(diabetes["insulin"])
```

```
# wine dataset
wine = pd.read_csv("wine.csv")
```

```
wine.head()
```

```
#histogram
fig, ax = plt.subplots(1,3, sharey=True, tight_layout=True)
ax[0].hist(wine['Alcohol'],bins=10, color='purple')
ax[1].hist(wine['Malic.acid'],bins=10, color='black')
ax[2].hist(wine['Phenols'],bins=10, color='blue')
```

```
#boxplot
plt.figure(figsize=(15,10))
boxplot_data = wine[['Alcohol', 'Malic.acid', 'Phenols']]
boxplot_data.boxplot()
```

```
# scatter plot
plt.figure(figsize=(15,10))
plt.scatter(wine['Alcohol'],wine['Hue'], marker='o', c='b')
plt.xlabel("Alcohol")
plt.ylabel("Hue")
plt.title("Correlation between Alcohol percent and Hue of wine")
plt.show()
```

```
#Violin plot
plt.violinplot(wine['Alcohol'])
```

```
# violin plot with seaborn
import seaborn as sns

sns.violinplot(x=wine['Wine'], y=wine['Alcohol'])
```

```
#line plot
plt.plot(wine["Alcohol"])
```
